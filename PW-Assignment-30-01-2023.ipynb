#### 4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

##### Answer:

```python
number = int(input("Enter a number to check:"))
count=0
while number > 10:
        number = number / 3
        count += 1
print("The number of times the number could be divided by 3 was", count) 
```

#### 5. Why and when to Use while Loop in Python give a detailed description with example.

###### Answer:-

A "while" loop in Python is a control flow structure used to repeatedly execute a block of code as long as a given condition is true. It continues to loop until the condition becomes false. The syntax for a while loop is as follows:

while condition:
    # code block to be executed
    
where condition is an expression that returns a boolean value (either True or False). The code block inside the while loop will be executed as long as the condition is True. Once the condition becomes False, the loop will terminate and the program will continue to execute the next line of code after the while loop.

In the above example, the condition count <= 5 is checked at the start of each iteration. If the condition is True, the loop will execute the code inside the loop and then increase the value of variable count by 1. The condition will be checked again and if it is still True, the loop will continue to repeat. This process continues until the condition becomes False, i.e., when count is more than 5.

The condition inside a while loop must eventually become False, otherwise the loop will run indefinitely and the program will get stuck in an infinite loop. To avoid this, it's a good practice to make sure that the value being used in the condition will change inside the loop so that the condition will eventually become False.

#### 6. Use nested while loop to print 3 different patterns

##### Answer:

```python
#Pattern 1
rows = 6
for i in range(1, rows + 1):
    j = 1
    while j <= i:
        print("*", end=" ")
        j = j + 1
    print()
    
```

```python
#Pattern 2
rows = 6
for i in range(rows, 0, -1):
    j = 1
    while j <= i:
        print("*", end=" ")
        j = j + 1
    print()
```

```python
#Pattern 3
rows = 5
for i in range(1, rows + 1):
    spaces = rows - i
    j = 1
    while j <= spaces:
        print(" ", end="")
        j = j + 1
    k = 1
    while k <= i:
        print("*", end=" ")
        k = k + 1
    print()
```
