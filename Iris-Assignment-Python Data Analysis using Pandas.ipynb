```python
#Ans22
def f(x):
    a=x[0]
    b=x[1]
    if (a =='Iris-setosa'):
        return b*2
    elif (a =='Iris-virginica'):
           return b*3
    else:
        return b*5
```

```python
iris["New_sepal_length"]=iris[["Species","SepalLengthCm"]].apply(lambda x:f(x),axis=1)
```
