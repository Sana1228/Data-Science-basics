#### Q4. What is DataFrame in pandas and how is it different from pandas.series? Explain with an example.

#### Answer:
In Pandas, a DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. It is a primary object in Pandas that is used to represent tabular data (similar to a spreadsheet or SQL table). A DataFrame is made up of one or more Pandas Series, where each column in a DataFrame represents a Series.

A Series, on the other hand, is a one-dimensional labeled array that can contain any data type (e.g. integers, strings, float values, etc.). Each element in a Series has a unique label or index that allows for easy identification and manipulation.

Here's an example to illustrate the difference between Pandas Series and DataFrame:

```python
import pandas as pd

# Create a Pandas Series
my_series = pd.Series([1, 2, 3, 4, 5, 6, 7])

# Create a Pandas DataFrame from a dictionary
data = {'Name': ['John', 'Emily', 'Luke', 'Sarah', 'Mark'],
        'Age': [24, 30, 18, 27, 42],
        'Gender': ['M', 'F', 'M', 'F', 'M']}
df = pd.DataFrame(data)

# Print the Series and DataFrame
print("Pandas Series:\n", my_series)
print("\nPandas DataFrame:\n", df)
```

In the output, you can see that the Pandas Series is a one-dimensional labeled array, whereas the Pandas DataFrame is a two-dimensional labeled data structure with columns of different data types. The DataFrame contains multiple Series (i.e. columns), each with its own unique label or index. The DataFrame can be thought of as a collection of Series, where each Series represents a column in the table.

#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?

#### Answer:
Pandas provides a wide range of functions to manipulate data in a DataFrame. Here are some of the most commonly used functions:

head() and tail(): These functions allow you to view the first or last few rows of a DataFrame, respectively. This is useful for quickly inspecting the structure and content of a DataFrame.

```python
print(df.head())  
print(df.tail())  
```

describe(): This function generates descriptive statistics for the numerical columns in a DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartiles.

```python
print(df.describe())
```

drop(): This function allows you to remove rows or columns from a DataFrame that you don't need.

```python
df = df.drop(['Gender'], axis=1)  # Remove two columns from the DataFrame
print(df.head())
```

#### Q7. Create a DataFrame using multiple Series. Explain with an example.

#### Answer:

```python
import pandas as pd

# Creating a series for student names
names = pd.Series(['John', 'Alice', 'Bob', 'Emily', 'Mike'])

# Creating a series for student ages
ages = pd.Series([18, 19, 17, 18, 20])

# Creating a series for student grades
grades = pd.Series([90, 85, 92, 88, 95])

# Combining the three series into a DataFrame
df = pd.DataFrame({'Name': names, 'Age': ages, 'Grade': grades})

# Displaying the DataFrame
print(df)
```
