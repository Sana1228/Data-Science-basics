```python
#Ans2
def calculate_age(birth_date):
    birth_date = datetime.strptime(birth_date, "%d-%m-%Y").date()
    till_date= '31-12-2021'
    till_date = datetime.strptime(till_date, "%d-%m-%Y").date()
    return till_date.year - birth_date.year - ((till_date.month, 
                                      till_date.day) < (birth_date.month, 
                                                    birth_date.day))
  
df['Age'] = df['DOB'].apply(calculate_age)
  
display(df)
```

```python
#Ans3
df2.merge(df[['Student_ID','Age']])
```
