```python
import pandas as pd
import numpy as np
import functools as ft
```

```python
# Task 1
# There are three different tables as given above. Please make three dataframes in python
# and save them as three .csv files. From Task 2 to Task 10, use the saved .csv files only.
```

```python
#Ans1
project_dataframe={'ID':['A001','A002','A003','A004','A005','A002','A005','A003','A001','A003','A001','A004','A004','A005'],
                  'Project':['Project 1','Project 2','Project 3','Project 4','Project 5','Project 6','Project 7','Project 8','Project 9','Project 10','Project 11','Project 12','Project 13','Project 14'],
                  'Cost':[1002000,2000000,4500000,5500000,'',680000,400000,350000,'',300000,2000000,1000000,3000000,200000],
                  'Status':['Finished','Ongoing','Finished','Ongoing','Finished','Failed','Finished','Failed','Ongoing','Finished','Failed','Ongoing','Finished','Finished']}
df1=pd.DataFrame(project_dataframe)
df1.to_csv('DF1.csv')
df1
```

```python
Employee_dataframe={'ID':['A001','A002','A003','A004','A005'],
                   'Name':['John Alter','Alice Luxumberg','Tom Sabestine','Nina Adgra','Amy Johny'],
                   'City':['Paris','London','Berlin','Newyork','Madrid'],
                   'Age':[25,27,29,31,30]}
df2=pd.DataFrame(Employee_dataframe)
df2.to_csv('DF2.csv')
df2
```

```python
Sld={'ID':['A001','A002','A003','A004','A005'],
    'Designation Level':[2,2,3,2,3]}
df3=pd.DataFrame(Sld)
df3.to_csv('DF3.csv')
df3
```

```python
# Task 2
# The cost column in dataframe “Project” has some missing values. Your task is to
# compute these missing values. Replace the missing values by running average. You
# should use “For” loop for this task.
```

```python
#Ans2
summ=0
count=0
for Cost in df1['Cost']:
    if(Cost!=""):
        count+=1
        summ+=Cost
avg=summ/count
df1 = df1.replace(r'^\s*$', avg, regex=True)
df1
```

```python
#Task 3
# Split the name column in Employee dataframe to two new columns “First Name”, “Last
# Name” and remove the older “name” column.
```

```python
#Ans3
# df2.loc[df2['Name'].str.split().str.len() == 2, 'last name'] = df2['Name'].str.split().str[-1]
df2[['First Name','Last Name']] = df2['Name'].loc[df2['Name'].str.split().str.len() == 2].str.split(expand=True)
df2.drop('Name', inplace=True, axis=1)
df2
```

```python
# Task 4
# Join all three dataframes in one single dataframe. Name it “Final”
```

```python
#Ans4
frames=[df1,df2,df3]
final = ft.reduce(lambda left, right: pd.merge(left, right, on='ID'), frames)
final
```

```python
# Task 5
# Add a new bonus column in Final dataframe. Give 5% bonus concerning project cost
# only for employees who have finished the projects.
```

```python
#Ans5
#Adding bonus of 5% of Cost to employess who have finished their project and 0 to those who haven't
final.loc[final['Status']=='Finished','Bonus']=(5/100)*final['Cost']
final.loc[final['Status']!='Finished','Bonus']=0
final
```

```python
# Task 6
# Decrease the designation level by 1, whose projects have status “fail”. Delete the
# employee record whose designation level is below 4. 
```

```python
#Ans6
final.loc[final['Status']=='Failed','Designation Level']-=1
final
#deleting records of employees whose designation level is below 4 as in greater than the value 4,since 2, 3 mid positions and 4 being least 
final1 = final.drop(final.index[final['Designation Level']>4])
final1
```

```python
# Task 7
# Add a new column for gender, than add “Mr.” and “Mrs” to the first name column.
```

```python
#Ans7
final1['Gender']=['Male','Male','Male','Female','Female','Male','Male','Male','Female','Female','Female','Female','Female','Female']
final1
final1.loc[final1['Gender']=='Male','First Name']='Mr. ' + final1['First Name']
final1.loc[final1['Gender']=='Female','First Name']='Mrs ' + final1['First Name']
final1
```

```python
# Task 8
# Increase designation level by 1 for the employees whose age is more than 29 years using
# IF condition
```

```python
#Ans8
final1.loc[final1['Age']>29,'Designation Level']+=1
final1
```

```python
# Task 9
# Add the cost of all projects for each Employee and save it in new dataframe
# “TotalProjCost” with three columns ID, First Name, and Total cost
```

```python
TotalProjCost=final1.groupby(['ID','First Name'])['Cost'].sum().reset_index()
```
