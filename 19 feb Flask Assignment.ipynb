#### Q1. What is Flask Framework? What are the advantages of Flask Framework?

#### Answer:

Flask is a micro web framework for building web applications using the Python programming language. It was created by Armin Ronacher in 2010 and is distributed as an open-source software under a BSD license. Flask is designed to be lightweight and modular, which makes it easy to learn and use for building small to medium-sized web applications.

Some of the advantages of using Flask for web development include:

1. Lightweight and modular: Flask is designed to be a minimalistic framework that is easy to use and understand. It doesn't come with a lot of built-in features and libraries, which makes it easy to customize and extend according to the specific needs of the project.

2. Flexible: Flask is a flexible framework that can be used to build a variety of web applications, from simple static websites to complex web applications with dynamic content and user authentication.

3. Easy to learn: Flask is one of the easiest web frameworks to learn, especially for developers who are already familiar with Python. The Flask documentation is well-written and provides a lot of examples, making it easy for developers to get started quickly.

4. Extensible: Flask can be extended with a wide range of third-party libraries and plugins, which can add features like authentication, database integration, and more.

5. Fast development: Flask makes it easy to build and deploy web applications quickly, which can save developers a lot of time and effort.

6. Large community: Flask has a large and active community of developers, which means that developers can easily find support, documentation, and third-party resources to help them build better web applications.

#### Q3. What is App routing in Flask? Why do we use app routes?

#### Answer:
In Flask, app routing is the process of mapping URLs to functions in the application. When a user visits a URL in a Flask application, the Flask app uses the app routing mechanism to determine which Python function to execute to handle the request.

App routing in Flask is done using the @app.route() decorator, which is used to associate a URL with a function in the application.

App routing is important in Flask because it allows developers to create web applications with multiple pages and functionalities. By using app routing, developers can map each URL to a specific function in the application, which makes it easy to manage and maintain the code.

In addition, app routing allows developers to create RESTful APIs, which are web services that follow the principles of Representational State Transfer (REST). RESTful APIs use HTTP methods like GET, POST, PUT, and DELETE to perform CRUD (Create, Read, Update, Delete) operations on data resources. With Flask's app routing mechanism, developers can easily create RESTful APIs by mapping URLs to functions that handle these HTTP methods.

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

#### Answer:
In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint and can also take parameters to generate dynamic URLs.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello people!'

@app.route('/user/<username>')
def user_profile(username):
    return f'Welcome, {username}!'

@app.route('/info')
def info():
    return 'Get the details here!.'

@app.route('/testimonials')
def testimonials():
    return 'This is the page for testimonials.'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('index')) 
        print(url_for('user_profile', username='Sana')) 
        print(url_for('info')) 
        print(url_for('testimonials'))  
```
